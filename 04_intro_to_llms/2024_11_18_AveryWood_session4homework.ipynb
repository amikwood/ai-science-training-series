```python
from transformers import AutoTokenizer,AutoModelForCausalLM, AutoConfig
input_text = ["My dog really wanted to eat icecream because",
              'What did they mean by that?',
              'Wait, do not',
              'Hey man, what',
              'Hey Jessica how are']

from transformers import pipeline
```

```python
generator = pipeline("text-generation", model="gpt2")
pipe = pipeline("text-generation", model="gpt2")
generator(input_text, max_length=20, num_return_sequences=5)
```

#Testing max length

```python
generator(input_text, max_length=30, num_return_sequences=5)
```

Increasing the max length actually increasead the diversity in sentence size for both queries and replies more than I was expecting. In terms of their legibility though that is still very hit or miss. Maybe you can argue that more context can be given with a larger max length.

#Testing temperature

```python
generator(input_text, max_length=20, num_return_sequences=5, temperature=0.7)
```

```python
generator(input_text, max_length=20, num_return_sequences=5, temperature=1)
```

```python
generator(input_text, max_length=20, num_return_sequences=5, temperature=0.1)
```

```python
generator(input_text, max_length=30, num_return_sequences=5, temperature=0.1)
```

Decreasing temperature definitely seems to lower the legibility of the generation. Most of the responses only seem to be half thoughts are the beginning of sentences. Increasing the max length here does seem to flesh out some of the sentences. Higher temperatures also seem to have more directly negative responses, with some very hurtful comments even thrown in.

#Bertviz attention mechanism visualization

```python
from transformers import AutoTokenizer, AutoModel, utils, AutoModelForCausalLM

from bertviz import model_view
utils.logging.set_verbosity_error()  # Suppress standard warnings

model_name = 'openai-community/gpt2'
input_text = "No, I am your father"
model = AutoModelForCausalLM.from_pretrained(model_name, output_attentions=True)
tokenizer = AutoTokenizer.from_pretrained(model_name)
inputs = tokenizer.encode(input_text, return_tensors='pt')  # Tokenize input text
outputs = model(inputs)  # Run model
attention = outputs[-1]  # Retrieve attention from model outputs
tokens = tokenizer.convert_ids_to_tokens(inputs[0])  # Convert input ids to token strings
model_view(attention, tokens)  # Display model view
```
